Libraries 

```python
import pandas as pd
import numpy as np
from collections import Counter
```

**The first dataset, “salesData,” contains information about medical devices that were sold to healthcare professionals (HCP) at various sites of care (SOC). Each object in the list [row in the table] represents a single invoice to a particular HCP at a particular SOC for a particular number of devices (volume).**


```python

df = pd.read_csv('salesData.csv')
df
```

**The second dataset, “procedureData,” contains information about medical procedures that were conducted by HCPs at various SOCs. Each object in the list [row in the table] represents some number (volume) of procedures that happened by a HCP at a SOC.
Let’s assume that the medical device is “disposable;” it can be used for one procedure and then it must be discarded.**

```python

df1 = pd.read_csv('procedureData.csv')
df1
```

**Finding rows in Procedure dataframe which are not available in Sales dataframe
[This will give us a dataset with HCP's who performed procedures at different SOC's not on the invoice of salesData.csv]**

```python


df2 = df1.merge(df, how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='left_only']
df2
```

**Manipulating dataset with sum of procedures(volume) for each HCP and storing it a new dataframe, Thus, creating a dataframe with hcp's and total number of procedures**




```python
df3 = df2.groupby(['hcp'])['volume'].sum().reset_index()
df3

#now we can see number of procedures done by HCP's (not on invoice) at all the different SOC's
```

**Top 10 HCPs who are performing the most procedures without the device**

```python

#finding max numer of procedures then its link with HCP using ID

a = df3['volume'].nlargest(n=10).reset_index()


hcp = []

for x in a['index']:
  hcp.append(df3['hcp'][x])

hcp
#Displaying HCP's with max procedures without the device
```

**Creating a DataFrame with Top 10 HCP and Total number of procedures without the device and displaying it**

```python

result = pd.DataFrame()
result['HCP'] = hcp
result['Total_Volume'] = a['volume']
result.index = pd.RangeIndex(start=1, stop=11, step=1)

result

#Our customers should focus on these HCP's 
```
